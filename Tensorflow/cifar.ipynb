<h1 align="center">Indice del analisis</h1>

Lo primero que vamos a hacer es importar las librerias y modulos necesarios para preparar los datos y entrenar el modelo:

```python

from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from tensorflow.keras import regularizers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D, Flatten, Dense, Dropout, Activation, BatchNormalization
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint
import numpy as np
import matplotlib.pyplot as plt 
from termcolor import colored
import warnings
warnings.filterwarnings("ignore")
```


<h1 align="center">Cargando dataset</h1>

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()
```

Los datos de imagenes son 50000 datos de 32 x 32 y de 3 canales (R,G,B). Son imagenes a color.

```python
x_train.shape
```

Ejemplo:

```python
plt.imshow(x_train[5])
```


<h1 align="center">Limpieza de datos</h1>

```python
x_train = x_train.astype('float32') / 255
x_test  = x_test.astype('float32')  / 255
```

```python
print('La longitud de la matrix es:', len(np.unique(y_train)))
```

Utilizamos to_categorical para convertir en vector la variable Y

```python
clases  = len(np.unique(y_train))
y_train = to_categorical(y_train, clases)
y_test  = to_categorical(y_test, clases)
```

Podemos observar que el output se convierte en vector:

```python
y_train[0]
```


<h1 align="center">Modelo Baseline convolucional</h1>

Para el entrenamiento de nuestro modelo Baseline utilizamos filtros, tamaño de filtros de 3x3, padding que nos mantenga la dimension de la imagen, capas relu como funcion de activacion y, por ultimo, flatten para llevar la ultima capa a una dimension y una capa densa con activacion softmax al ser un modelo de clasificacion multiple.

```python
base_filtros  = x_train.shape[1:][0]
w_regulatizer = 1e-4
```

```python
model = Sequential()

### Capa numero 1: ###
model.add(Conv2D(filters     = base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer), # Regularizacion L2
                 input_shape = x_train.shape[1:])) # Va a recibir imagenes de 32x32 en 3 distintas capas

model.add(Activation('relu')) # Funcion de activacion Relu



### Capa numero 2: ###
model.add(Conv2D(filters     = base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu



### Capa numero 3: ###
model.add(Conv2D(filters     = 2 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu



### Capa numero 4: ###
model.add(Conv2D(filters     = 2 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu




### Clasificacion - Flatten ###

model.add(Flatten()) # Llevamos todo el resultado de la convolucion 6 a una sola dimension
model.add(Dense(clases, # Son 10 neuronas para el numero de clases
                activation='softmax' # softmax al ser clasificacion multiple
               ))
```


<h1 align="center">Neruonal Net</h1>

```python
def _draw_neural_net(ax, left, right, bottom, top, layer_sizes):
        n_layers = len(layer_sizes)
        v_spacing = (top - bottom)/float(max(layer_sizes))
        h_spacing = (right - left)/float(len(layer_sizes) - 1)
        # Nodes
        for n, layer_size in enumerate(layer_sizes):
            layer_top = v_spacing*(layer_size - 1)/2. + (top + bottom)/2.
            for m in range(0 , layer_size + 1):
                circle = plt.Circle((n*h_spacing + left, 
                                     layer_top - m*v_spacing), 
                                    v_spacing/4.,
                                    color='w', 
                                    ec='k', 
                                    zorder=3)
                ax.legend([circle], ['Inputs - Hidden Labels - Outputs'], loc='upper left')
                ax.add_artist(circle)
        # Edges
        for n, (layer_size_a, layer_size_b) in enumerate(zip(layer_sizes[:-1], layer_sizes[1:])):
            layer_top_a = v_spacing*(layer_size_a - 1)/2. + (top + bottom)/2.
            layer_top_b = v_spacing*(layer_size_b - 1)/2. + (top + bottom)/2.
            
            for m in range(0 ,layer_size_a +1):

                for o in range(0, layer_size_b + 1):
                    line = plt.Line2D([n*h_spacing + left, (n + 1)*h_spacing + left],
                                    [layer_top_a - m*v_spacing, layer_top_b - o*v_spacing], 
                                      c='b', alpha=0.3)
                    ax.add_artist(line)
```

Compliacion del modelo

```python
from tensorflow.keras import optimizers

model.compile(loss = 'categorical_crossentropy', 
              optimizer = optimizers.Adam(),
              metrics = ['accuracy'])
```

Entrenamiento del modelo:

batch size: En cada capa entrenamos los datos por pequeños lotes de a 32. Si hay muchas imagenes es importante este parametro ya que entre menos sea el batch size menos recursos utilizara el modelo.

epochs: cantidad de iteraciones del modelo

validation_data: set de validacion

verbose: el modo en el que se van a ver los datos

shuffle: mezla de datos y clases

```python
try:
    if modelo.lower() == 'si':
        hist = model.fit(x_train, y_train, 
                         batch_size = tamaño_batch,
                         epochs     = epocas,
                         validation_data = (x_test, y_test), 
                         verbose = True, 
                         shuffle = True)
    else:
        print(colored('\nVer la pregunta de la celda de arriba\n', 'red'))
except:
    print('No se definio modelo')
```

<h1 align="center">Graficos</h1>

```python
def _get_plots(metricas_train, metricas_test, hyper_p, y_label = 'Metrics'):
    capas = hyper_p['capas']
    epocas = hyper_p['epocas']
    plt.figure(figsize = (8,5))
    plt.plot(
            metricas_train,
            color = 'blue', 
            linestyle = 'dashed', 
            marker = '*', 
            markerfacecolor = 'black', 
            markersize = 5,
            label = 'Training')
    plt.plot(
            metricas_test,
            color = 'red', 
            linestyle = 'dashed', 
            marker = '*', 
            markerfacecolor = 'black', 
            markersize = 5,
            label = 'Test')
    plt.title(f'Datos {y_label}', fontsize=16)
    plt.suptitle(f'Capas: {capas}, Epocas: {epocas}', fontsize=16)
    plt.xlabel('Epocas', fontsize=16); plt.ylabel(y_label, fontsize=16)
    leg = plt.legend(loc='upper left')
    return plt.grid(); plt.show()
```


<h1 align="center">Data augmentation</h1>




<h3 align="center">¿Como mejorar nuestro modelo?</h3>

La clasificacion de imagenes no suele ser sencilla, tomamos multiples fotos desde distintos focos de luz, distintas formas y luego clasificarlos. Para mejorar las metricas del modelo podemos utilizar Data Augmentation que consiste en hacer rotaciones, aumentar el brillo de la imagen y otras configuraciones que podemos hacer manteniendo la imagen y sus caracteristicas principales pero alterando un poco estas mismas.

Rotacion: Podemos configurar en keras una rotacion y en donde queden puntos negros que complete los mismos con pixeles cercanos.

Desplazamiento: Desplazar la imagen a los costados o hacia arriba/abajo, como queramos desplazar la imagen conservando los principales atributos.

Brillo: Esto va a cambiar el mapa de los pixeles pero siempre conservando los principales atributos.

zoom: Otra de las cosas que podemos alterar.


Data Augmentation tiene mucho valor al entrenar nuestra red convolucional, podemos tener mas set de entrenamiento para que nuestro modelo aprenda mucho mejor.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
datagen = ImageDataGenerator(
          rotation_range  = 15,    # Grados de Rotacion
          width_shift_range = 0.1, # Cambio de posicion de la Imagen
          height_shift_range= 0.1, # Cambio de posicion de la Imagen
          zoom_range = 0.1,        # Cambio de posicion de la Imagen
          horizontal_flip = True,  # Cambio espejo en el eje horizontal
          vertical_flip = True,    # Cambio espejo en el eje vertical
          brightness_range = [0.4, 1.5])
```

<h1 align="center">Early stopping</h1>




Early stopping es un proceso que detiene el entrenamiento de la Red Neuronal cuando una de las metricas seleccionadas no tiene mas performance, es decir, no mejora por mas que se incrementen las epocas.
Se puede poner como variable la metrica y la cantidad de epocas maximas.

```python
ear = keras.callbacks.EarlyStopping(monitor = 'accuracy',
                                    patience = 3) # Esta metrica es importante ya que si es muy corto el entrenamiento se cortara rapido
```

<h1 align="center">Batch Normalization</h1>

Dentro de las capas ocultas se pueden hacer normalizaciones en cada una de ellas a lo largo del entrenamiento. Los pesos estan variando constantemente en el transcurso de las epocas a lo largo de la Red Neuronal, estos valores se pueden estandarizar dentro de las mismas capas ocultas. Este proceso se llama Batch Normalization.

<h1 align="center">DropOut y Pooling</h1>

DropOut y Pooling son hyperparametros que se pueden adicionar al modelo para reducir overfitting. 

DropOput consiste en desactivar neuronas en cada capa, es decir, un dropoup del 50 % desactivara la mitad de las neuronas de una de las capas, este proceso ayuda a recudir el sobreajuste.

Otro de los metodos es el Pooling, este mismo consiste en una matriz que se utiliza como filtro y saca el maximo valor o promedio de una parte de la imagen.

Un ejemplo seria el siguiente:

<b> Si la matriz de filtro es de 2x2 extrae el mayor valor de cada parte de la matriz original y crea una nueva de 2x2. Este proceso disminuye el ancho y largo de nuestra imagen original pero nos ayuda a reducir la complejidad del modelo. </b>

<h1 align="center">Segundo Modelo</h1>

Para entrenar nuestro segundo modelo se agregaron los siguientes parametros, hyperparametros y transformacion de datos:

- Normalizacion
- 2 Nuevas Capas Ocultas
- Pooling
- Dropout
- Batch Normalization
- Early Stopping
- Data augmentation
- Learning Rate

```python
(x_train, y_train), (x_test, y_test) = cifar10.load_data()


x_train = x_train.astype('float32') 
x_test  = x_test.astype('float32') 


clases  = len(np.unique(y_train))
y_train = to_categorical(y_train, clases)
y_test  = to_categorical(y_test, clases)


# Normalizacion

x_train = (x_train - np.mean(x_train)) / (np.std(x_train) + 1e-7)
x_test  = (x_test - np.mean(x_test)) / (np.std(x_test) + 1e-7)

base_filtros  = x_train.shape[1:][0]
w_regulatizer = 1e-4
```

```python
model = Sequential()

### Capa numero 1: ###
model.add(Conv2D(filters     = base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer), # Regularizacion L2
                 input_shape = x_train.shape[1:])) # Va a recibir imagenes de 32x32 en 3 distintas capas

model.add(Activation('relu')) # Funcion de activacion Relu
model.add(BatchNormalization())


### Capa numero 2: ###
model.add(Conv2D(filters     = base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu
model.add(BatchNormalization())

model.add(MaxPooling2D(pool_size = (2,2))) # Reduccion de complejidad del Modelo (Tamaño del pool de 2x2)

model.add(Dropout(0.1)) # Dropoutl del 10 % para reducir overfitting



### Capa numero 3: ###
model.add(Conv2D(filters     = 2 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu
model.add(BatchNormalization())

model.add(Dropout(0.15)) # Dropoutl del 15 % para reducir overfitting



### Capa numero 4: ###
model.add(Conv2D(filters     = 2 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu
model.add(BatchNormalization())

model.add(MaxPooling2D(pool_size = (2,2))) # Reduccion de complejidad del Modelo (Tamaño del pool de 2x2)

model.add(Dropout(0.25)) # Dropoutl del 25 % para reducir overfitting




### Capa numero 5: ###
model.add(Conv2D(filters     = 4 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2


model.add(Activation('relu'))
model.add(BatchNormalization())


### Capa numero 6: ###
model.add(Conv2D(filters     = 4 * base_filtros, # (Cantidad de filtros de la capa)
                 kernel_size = (3,3), # (Tamaño del Kernel)
                 padding     ='same', # (same mantiene la misma dimension de la imagen de entrada)
                 kernel_regularizer = regularizers.l2(w_regulatizer))) # Regularizacion L2

model.add(Activation('relu')) # Funcion de activacion Relu
model.add(BatchNormalization())

model.add(MaxPooling2D(pool_size = (2,2))) # Reduccion de complejidad del Modelo (Tamaño del pool de 2x2)

model.add(Dropout(0.35)) # Dropoutl del 35 % para reducir overfitting


### Clasificacion - Flatten ###

model.add(Flatten()) # Llevamos todo el resultado de la convolucion 6 a una sola dimension
model.add(Dense(clases, # Son 10 neuronas para el numero de clases
                activation='softmax' # softmax al ser clasificacion multiple
               ))

opt = keras.optimizers.RMSprop(learning_rate = 0.0001, 
                               decay = 1e-6)

model.compile(loss = 'categorical_crossentropy', 
              optimizer = opt,
              metrics = ['accuracy'])


model.summary()
```

<h1 align="center">Entrenando el Modelo</h1>

```python
epocas       = 30
tamaño_batch = 32

hist = model.fit(datagen.flow(x_train, y_train, batch_size = tamaño_batch),
                             epochs     = epocas,
                             validation_data = (x_test, y_test), 
                             verbose = True, 
                             shuffle = True,
                             steps_per_epoch = x_train.shape[0] // tamaño_batch,
                             callbacks = [ear])
```

<h1 align="center">Resultados</h1>

```python
_get_plots(hist.history['accuracy'],
               hist.history['val_accuracy'],
               hyper_p = { 'capas'  : 6, 'epocas' : 30},
               y_label = 'accuracy')

_get_plots(hist.history['loss'],
           hist.history['val_loss'],
           hyper_p = {'capas'  : 6, 'epocas' : 30},
           y_label = 'loss')
```

```python
metricas = model.evaluate(x_test, y_test)

print('Loss:     ',metricas[0])
print('Accuracy: ', metricas[1])
```
