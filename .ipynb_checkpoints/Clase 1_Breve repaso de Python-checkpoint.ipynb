*   **List**

Una lista es una secuencia de elementos. Las listas permiten combinar elementos de diferente tipo. Son mutables, es decir se puede cambiar su contenido, por lo cuál dan flexibilidad al manipular objetos. Una lista se crea usando corchetes y separando los elementos con coma.

```
lista=['gato', 1,True,5E2] # una lista puede contener objetos de distinto tipo
lista
```

*   **Tuples**

Una tupla es un objeto secuencial como las listas, pero no son mutables.Una tupla se crea usando paréntesis  y separando los elementos con coma.

```
tupla=('gato', 1,True,5E2) # una tupla puede contener objetos de distinto tipo
tupla
```

```
type(tupla)
```

```
#Se accede a los elementos de una tupla con corchetes indicando la posición, que comienza en 0
tupla[0]
```

```
# Veamos que no son mutables, es decir, no se pueden modificar sus elementos
tupla=('gato', 1,True,5E2)
print('tupla vieja ',tupla)
tupla[0]=5 #dará error 

```

```
print('tupla nueva:', tupla) #imprime la tupla definida anteriormente
```

*   **Diccionarie**

Un diccionario es un objeto que permite asociar claves a valores.
Es decir nos permiten almacenar una serie de mapeos entre dos conjuntos de elementos, llamados keys and values (Claves y Valores). Son mutables, es decir se puede cambiar su contenido. Un diccionario se crea usando llaves y separando los elementos con coma.


```
diccionario = {'Nombre':'Juan','Edad':30}  # Cada elemento en un diccionario contiene una clave y un valor - es decir un par de clave-valor
print(diccionario) # Cada par de clave-valor es denominado como elemento (item).
```

```
diccionario['Nombre'] #como el índice son las claves,  nos devuelve el valor correspondiente
```

```
#podemos cambiar un elemento del diccionario usando la llave
diccionario['Edad']=40
print(diccionario)
```

```
# podemos agregar un nuevo elemento al diccionario previamente definido
diccionario['Altura']=1.72
print(diccionario)
```

# 3. Librerías en Python

#### ¿Qué es una librería?

En informática, una biblioteca o, llamada por vicio del lenguaje librería (del inglés library) es un conjunto de implementaciones funciones, codificadas en un lenguaje de programación, que realizan una serie de tareas específicas y se incluyen o importan en el programa evitando tener que reescribir esos códigos.


**Importar librerias**
** Python posee ya muchas librerías con funciones específicas según la tarea que querramos realizar. Existen librerías de cálculo, de visulización, de estadística, procesamiento de señales, de algebra lineal y de redes neuronales artificiales (entre muchas mas). 

La forma de "llamar", "cargar" o importar una libreria xxxxxx es la siguiente: 
* import xxxxxx as xx 
* donde xx es el nombre corto que le asignamos para poder llamar sus funciones.


Veamos algunas de las librerías más utilizadas en Python

# Librerías Matplotlib y Seaborn

Son dos librerías de visualización. Matplotlib provee funciones para crear y editar gráficas enriquecidas que se adaptan a distintos tipos de conjunto de datos. 
Seaborn está desarrollada sobre matplotlib. Ofrece una interfaz de alto nivel para la creación  gráficas más atractivas. Se encuentra además integrada con las estructuras de datos de pandas.

```
# importamos la libreria matplotlib
import matplotlib.pyplot as plt
```

```
#Hacemos un gráfico de barras para los datos del Titanic
Datos.Survived.value_counts().plot(kind="bar", alpha=0.5) 
plt.title("Sobrevivieron - cuenta total -")
```

```
# Hacemos  barras horizontales y editamos el gráfico
fig = plt.figure(figsize=(30,10))
Datos.Sex[Datos.Survived == 1].value_counts(normalize = True).plot(kind="barh", alpha=0.5)
plt.title("Sobrevivientes - Male vS female")
plt.show()
```

```
import seaborn as sns #Se importa con matplotib y pandas, que ya están importadas
```

```
# Barras adyacentes
sns.countplot(x='Sex', hue='Survived', data=Datos)
plt.title('Supervivencia desagregada por genero')
plt.show()
```

```
# Histograma segun la edad
Datos.hist('Age')
```

```
# Distribución del valor  la tarifa por clase
Datos.boxplot('Fare', 'Pclass', figsize=(6, 4), grid=False)
```

```
#Otra forma
sns.boxplot( y=Datos["Fare"], x=Datos["Pclass"] )
plt.show()
#https://pythonbasics.org/seaborn-boxplot/
```

```
sns.violinplot(data = Datos, y = "Fare", x = "Pclass")
```
