**Transformación de variables**

```python
# Transformación de variable education
def convert_education(a):

    if a == 'primary' :
        return 1
    elif a == 'secondary':
        return 2
    elif a == 'tertiary' :
        return 3
    elif a == 'unknown':
        return 0
    else:
        return a

df["education"] = df["education"].apply(convert_education)

# Transformación de variable month
def get_month(x):
    month =  { 'jan': 1,
            'feb': 2,
            'mar': 3,
            'apr': 4,
            'may': 5,
            'jun': 6,
            'jul': 7,
            'aug': 8,
            'sep': 9, 
            'oct': 10,
            'nov': 11,
            'dec': 12}
    return month.get(x)

df["month"] = df["month"].apply(get_month)

# Función de transformación de variables dicotómicas
def convert_dicot(a):

    if a == 'yes' :
        return 1
    else:
        return 0

```

```python
# Se transformarán en dicotómicas (Yes = 1 , No = 0) las variables: default, housing, loan.

# Se transforma default
df['default'] = df['default'].apply(convert_dicot)

# Se transforma housing
df['housing'] = df['housing'].apply(convert_dicot)

# Se transforma loan
df['loan'] = df['loan'].apply(convert_dicot)

# Se transforma loan
df['y'] = df['y'].apply(convert_dicot)
```

**Se estudian outliers**

Se define una función para estudiar todos los outliers. 

```python
def outlier(df_in, col_name):

    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr

    print("IQR Limite inferior "+str(fence_low))
    print("IQR Limite superior "+str(fence_high))
    print("IQR = ", iqr)

    df_filtered = df_in.loc[(df_in[col_name] > fence_low) & (df_in[col_name] < fence_high)]
    df_outliers = df_in.loc[(df_in[col_name] < fence_low) | (df_in[col_name] > fence_high)]

    return df_filtered, df_outliers, fence_low, fence_high, col_name


```

```python
campos = ['age',
'balance',
'day',
'duration',
'campaign',
'pdays',
'previous']


lista_df_outliers = []

for campo_outlier in campos:
    df_filtered, df_outliers, superior, inferior, nombre_campo = outlier(df,campo_outlier)
    lista_df_outliers.append([campo_outlier,df_filtered, df_outliers, superior, inferior,nombre_campo])


for df_outliers_aux in lista_df_outliers:
    if len(df_outliers_aux[2]) > 0:
        print("-----------------------------------")
        print(df_outliers_aux[5])
        print(df_outliers_aux[0])
        print(df_outliers_aux[2][df_outliers_aux[0]])
        print("inferior")
        print(df_outliers_aux[3])
        print("superior")
        print(df_outliers_aux[4])



```

```python
# Se define función para la exploración visual 
def expl_visual(df_histograma,field):

  f, (ax_box, ax_hist) = plt.subplots(2, sharex=True, gridspec_kw= {"height_ratios": (0.2, 1)})
  mean=df_histograma[field].mean()
  median=df_histograma[field].median()
  mode=df_histograma[field].mode().values[0]

  sns.boxplot(data=df_histograma, x=field, ax=ax_box)
  ax_box.axvline(mean, color='r', linestyle='--')
  ax_box.axvline(median, color='g', linestyle='-')
  ax_box.axvline(mode, color='b', linestyle='-')

  sns.histplot(data=df_histograma, x=field, ax=ax_hist, kde=True)
  ax_hist.axvline(mean, color='r', linestyle='--', label="Mean")
  ax_hist.axvline(median, color='g', linestyle='-', label="Median")
  ax_hist.axvline(mode, color='b', linestyle='-', label="Mode")

  ax_hist.legend()

  ax_box.set(xlabel='')
  plt.show()

  print('Media = ',mean)
  print('Mediana = ',median)
  print('Moda principal = ',mode)
```

```python
# Analisis Bivariado para cualitativas CHI CUADRADO (entre variables) sobre el dataframe df  

#Devuelve solo aquellos que hayan tenido p_value menor a 0.05
def chi_cuadrado(df,field1,field2):

    from scipy import stats
    #stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
    
    dep1 = pd.crosstab(df[field1],df[field2])

    chi2, p, dof, expected = stats.chi2_contingency(dep1)

    dep1
    if p < 0.05:
        print(field1)
        print(field2)
        print('Chi-squared Test Statistic: %s' % chi2)
        print('P-value: %s' % p)
        print('Degrees of Freedom: %s\n' % dof)

#Recorre la lista de campos para calcular chi cuadrado
def show_chi2(lista_chi2):

    i=0
    j=0

    for campo1 in lista_chi2:
        j = 0
        i = i + 1
        for campo2 in lista_chi2:
            j = j + 1 
            if j > i:
                chi_cuadrado(df,campo1,campo2)



#Se cargan los campos para validar si tienen relacion
lista_chi2 = ['education',
              'default', 
              'housing',
              'loan',
              'day',
              'month',
              'job',
              'marital',
              'contact',
              'poutcome']

show_chi2(lista_chi2)
```

##**3- DATA MODELING** 

El objetivo del presente trabajo es predecir qué personas se convertirán en clientes del banco. Por ello, en esta fase de modelado consideraremos diversos MODELOS PREDICTIVOS.

Debido a que el dataset proporciona la variable "y" que indica en base al historial qué personas se han convertido en clientes, esa será la variable TARGET que funcionará como etiqueta. En consecuencia, se optará por modelos supervisados.  

La variable Target es dicotómica (0-1) por lo que se busca predecir un valor discreto (cancelado-no cancelado). Por lo tanto, el Analytical Approach será una TAREA DE CLASIFICACION.


Para llevar adelante los modelos se generarán sets de entrenamiento (train) y prueba (test). También se utilizarán las validaciones cruzadas (K-fold) a fin de evitar el sobreajuste de los modelos. Asimismo, a fin de balancear el dataset se establecerá en los algoritmos el parámetro class_weight = 'balanced' a fin de calcular y ajustar automáticamente los pesos inversamente proporcionales a las frecuencias de clase en los datos de entrada.

Diversos algoritmos de clasificación pueden ser utilizados para el modelado del caso de negocio planteado: Regresión Logística, KNN, Naive Bayes, Árboles de decisión, Random forest, Redes Neuronales.

En este trabajo se desarrollarán sobre la base de los dos dataframes específicamente tres algoritmos:

##**CONCLUSIONES FINALES** 

**Evaluación del modelo a la luz del caso de negocio**

Para evaluar el modelo a elegir resulta necesario establecer los criterios de aceptabilidad del mismo. Si el modelo acierta, tanto en 'posibles clientes' (VP) como en 'posibles no clientes' (VN) sería beneficioso para el banco, ya que conllevaría una captación de clientes y evitaría el gastos operativos respectivamente. 

Un problema radica en los errores que podría generar el modelo:

+ Error I - Falso Positivo: El modelo predice cliente y se llama, pero no se logra. Se incurre en un gasto operativo que impactan en los costos de la campaña de marketing, sin embargo los costos suelen estar contemplados a priori y no se lograría una reducción de los mismos. 
+ Error II - Falso Negativo: El modelo predice que no es un posible cliente por lo tanto no se lo contacta. Sin embargo, lo sería. La agencia no pierde, pero no gana. 

Sería relevante poder definir estos criterios de manera informada y junto con el stakeholder durante la fase de Bussiness Understanding. Considerando que el caso de negocio se trata de una campaña de marketing para captación de clientes, se consideraría más desfavorables un error tipo II que un error tipo I. Aunque ninguno de estos errores conllevaría consecuencias graves en sí mismas. 

Por ello, resulta indispensable comparar las métricas de evaluación de los tres algoritmos a la luz de lo mencionado: donde la métrica de interés principal son los Verdaderos Positivos(↑) y resuta relevante considerar los menores casos de Falsos Negativos(↓). De esta manera, se logrará optimizar la captación de más clientes. 

Si bien los modelos de Random Forest y Redes Neuronales han reportado mejores valores de Exactitud y Precisión, se observan bajos niveles de Sensibilidad. Para esta última métrica quien se destacó fue la Regresión Logística. 
El caso de negocio requiere en sí mismo una optimización en la captación, siendo el punto de interés principal (que estimamos estaría consensuado con el cliente). Dadas las características del caso, considerar la métrica de SENSIBILIDAD resulta la más importante en esta ocasión, por encima de la precisión y exactitud del modelo. 

En consecuencia, resulta totalmente relevante que el modelo pueda detectar correctamente los verdaderos positivos (el modelo predice que serán clientes y efectivamente se logra captar) y cometer menor cantidad de errores tipo II (el modelo predice que no será cliente pero la persona se convierte en cliente). En este sentido, se puede observar que el Modelo de Regresión Logística Nº2 presenta los mejores valores. 

En congruencia, la sensibilidad del Modelo Nº2 presenta mejores valores identificando en el 61% de los casos. Por lo tanto, se concluye que el modelo que presenta mejor desempeño resulta ser el **MODELO DE REGRESIÓN LOGÍSTICA Nº2** basado en el df_2 con las variables:
