```
from datetime import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
```

El archivo Housing.txt 
Corresponde a los datos de comienzo (starts) y final (comps) de obra para los EEUU para el per√≠odo Enero 1968 a Junio 1996. Los datos tienen frecuencia mensual y ya se encuentran ajustado estacionalmente.


```
df = pd.read_csv('Housing.txt') 
for c in ['starts', 'comps']:
   df[c] = (df[c] - df[c].min()) / (df[c].max() - df[c].min())

df.head()
```

```
for i, c in enumerate(df.columns.drop(['Unnamed: 0','month'])):
    plt.figure(figsize=(20, 20))
    plt.subplot(df.columns.size, 1, i+1)
    plt.plot(df['month'], df[c])    
    plt.title(c, y=0.8, loc='right') 
     
    plt.show()
```

Realizo en un mismo gr√°fico las dos series para comparar su comportamiento

```
plt.figure(figsize=(15, 5))
plt.plot(df['month'], df['starts'], label="Obras iniciadas")
plt.plot(df['month'], df['comps'], label="Obras finalizadas")
plt.xlabel("Meses")
plt.ylabel("Obras")
plt.legend()
plt.show()
```

"Unnamed" es una especie de "contador secuencial" de los datos ordenados, por eso se usar√° para seleccionar el conjunto de entrenamiento los primeros 20 a√±os (70% training y 30% testing) 

```
train = df[df['Unnamed: 0'] < 241]
test  = df[df['Unnamed: 0'] >= 241]
train.shape, test.shape
```

Predicci√≥n para "starts"

```
window_size = 1
cols = df.columns.drop(['Unnamed: 0', 'month'])
X_train = pd.concat([ train[c].shift(1, fill_value=0) for c in cols ], axis=1, keys=cols)
y_train = train['starts']
X_test  = pd.concat([ test[c].shift(1, fill_value=0) for c in cols ], axis=1, keys=cols)
y_test  = test['starts']
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```
X_train = X_train.values.reshape( (X_train.shape[0], X_train.shape[1], window_size) )
X_test  = X_test.values.reshape( (X_test.shape[0], X_train.shape[1], window_size) )
X_train.shape, X_test.shape
```

```
model = keras.models.Sequential([
    keras.layers.Input(shape=[X_train.shape[1], X_train.shape[2]]),
    keras.layers.LSTM(120),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(1, activation='linear'),
 ])
    
model.compile(loss="mean_squared_error", optimizer="adam")
history = model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=35, batch_size=50, verbose=True)
```

```
pred = model.predict(X_test)
```

```
# RMSE
np.sqrt(np.mean((pred.flatten() - y_test)**2))
```

```
plt.figure(figsize=(15, 5))
plt.plot(df['month'], df['starts'], label="Obras Iniciadas")
plt.plot(test['month'][:-window_size], pred[window_size:], label="prediccion")
plt.xlabel("meses")
plt.ylabel("Obras")
plt.legend()
plt.show()
```
