**Clase 5: Ejemplo de LSTM para series temporales**

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tensorflow import keras
```

```
#https://www.kaggle.com/datasets/andreazzini/international-airline-passengers

df = pd.read_csv('/content/international-airline-passengers.csv')
df.columns = ['yyyy-mm', 'Pasajeros']
df.head()
```

```
df.shape
```

```
df
```

```
df = df.iloc[:-1] #eliminamos la ltima fila
```

```
df 
```

```
plt.figure(figsize=(15, 5))
plt.plot(df['yyyy-mm'], df['Pasajeros'], label="Pasajeros areos")
plt.xlabel("Meses")
plt.ylabel("Pasajeros")
plt.xticks(rotation = 90) #roto las etiquetas
plt.legend()
plt.show()
```

```
#from sklearn.preprocessing import MinMaxScaler #para reescalar 
#MinMaxScaler(feature_range=(0,1))
```

```
# Reescalamiento a 0-1 
#Para no "perder" la variable Pasajeros
#definimos una nueva variable escalada
df['Pasajeros01'] = (df['Pasajeros'] - df['Pasajeros'].min()) / (df['Pasajeros'].max() - df['Pasajeros'].min())
df.head()
```

```
#Separamos en training y testing 
train = df[df['yyyy-mm'] < '1958-01']
test  = df[df['yyyy-mm'] >= '1958-01']
train.shape, test.shape
```

```
#Trabajamos con la variable escalada
window_size = 1 #Tomamos ventana de un mes 
X_train = pd.concat([train['Pasajeros01'].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"Pasajeros01_{w}" for w in range(1, window_size+1)])
y_train = train['Pasajeros01']
X_test  = pd.concat([test['Pasajeros01'].shift(w, fill_value=0) for w in range(1, window_size+1)], axis=1, keys=[f"Pasajeros01_{w}" for w in range(1, window_size+1)])
y_test  = test['Pasajeros01']
```

Comentario: dataframe.shift() es una funcin de pandas. ES un ndice de desplazamiento para un nmero determinado de perodos con una frecuencia de tiempo opcional. Esta funcin toma un parmetro escalar llamado perodo, que representa el nmero de desplazamientos a realizar sobre el eje deseado. Es una funcin muy til cuando se trata de datos de series de tiempo.

```
#Estamos tomando solo el mes anterior
X_train[:7], y_train[:7]
```

```
 #Definimos la arquitectura 
 model = keras.models.Sequential([
    keras.layers.Input(shape=[1, window_size]), #capa de entrada : Entra 1 dato con informacin de 1 mes
    keras.layers.LSTM(4), #Este valor es cuanto quiero que recuerde
    keras.layers.Dense(1),#Capa de salida:solo quiero predecir un valor
 ])
model.summary()
```

```
#Compilamos y ejecutamos
model.compile(loss="mean_squared_error", optimizer="adam", metrics="mean_squared_error")
model.fit(X_train, y_train, epochs=100, batch_size=1, verbose=True)
```

```
pred = model.predict(X_test)
#print(pred)
pred.shape
```

```
X_test
```

```
#pred
pred.flatten()
pred.flatten().shape
```

```
# RMSE
np.sqrt(np.mean((pred.flatten() - y_test)**2))
```

```
#Otra forma de obtener RMSE
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(pred.flatten(), y_test)
RMSE = np.sqrt(MSE)
print("Root Mean Square Error:",RMSE)
```

```
#Grafico con la variable escalada 
plt.figure(figsize=(15, 5))
plt.plot(df['yyyy-mm'], df['Pasajeros01'], label="Pasajeros areos")
plt.plot(test['yyyy-mm'][:-window_size], pred[window_size:], label="Prediccion")
plt.xlabel("Meses")
plt.ylabel("Pasajeros01")
plt.legend()
plt.show()
```

```
#En las unidades originales
#df['Pasajeros01'] = (df['Pasajeros'] - df['Pasajeros'].min()) / (df['Pasajeros'].max() - df['Pasajeros'].min())  # Escalamos
predorig=pred*(df['Pasajeros'].max() - df['Pasajeros'].min())+df['Pasajeros'].min() # Pasamos las predicciones a las unidades originales 

predorig[:3] #inspeccionamos las primeras 3
```

```
#Para graficar en las unidades originales
#Observamos el mismo grfico, con escala diferente en el eje y.
plt.figure(figsize=(15, 5))
plt.plot(df['yyyy-mm'], df['Pasajeros'], label="Pasajeros areos")
plt.plot(test['yyyy-mm'][:-window_size], predorig[window_size:], label="Prediccion") #Usamos predorig
plt.xlabel("Meses")
plt.ylabel("Pasajeros")
plt.legend()
plt.show()
```
